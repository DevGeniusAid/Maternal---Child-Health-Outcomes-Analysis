```python
# df1
#Convert all object type columns in your DataFrame to category
df1 = df1.astype({col: 'category' for col in df1.select_dtypes(include='object').columns})

# Parse dates separately
parse_dates = ['START', 'END', 'DOB', 'FATHER_DEATH_DATE', 'MOTHER_DEATH_DATE']

# Apply dtypes and parse_dates to df1
df1[parse_dates] = df1[parse_dates].apply(pd.to_datetime, errors='coerce') 
```
